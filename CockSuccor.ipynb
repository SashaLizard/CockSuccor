Setup

```python
from os import getcwd, getenv
from io import BytesIO
import PySimpleGUI as ui
from selenium.webdriver import Firefox
from selenium.webdriver.chrome.service import Service
from webdriver_manager.firefox import GeckoDriverManager
from requests import Session
from PIL import Image
from bs4 import BeautifulSoup as bs
from pathlib import Path

cwd = getcwd()

cock_path = getenv('APPDATA')[:-7] + 'Local\Cockatrice\Cockatrice\pics\CUSTOM'
Path(cwd + "/image_dump").mkdir(parents=True, exist_ok=True)
Path(cwd + "/website_dump").mkdir(parents=True, exist_ok=True)
```

Scraping

```python
driver = Firefox(service=Service(GeckoDriverManager().install()))
session = Session()

index = 1

def get_img_sml(cnt, url):

    img_list = []

    for i in range(cnt):
        img = Image.open(BytesIO(session.get(url[i]).content))
        img_path = f'image_dump\\{i}.png'
        img_list.append(img_path)
        img.resize((168, 234)).save(f'image_dump\\{i}.png')

    global window
    if current_page != 1:
        window[f'BACK'].update(f'Page {current_page - 1}', visible=True)
    if cnt >= 60:
        window[f'NEXT'].update(f'Page {current_page + 1}', visible=True)

    return img_list

def get_img_big(url):

    img = Image.open(BytesIO(session.get(url).content))
    img_path = f'image_dump\\biggus.png'
    img_resized = img.resize((int(168*2.5), int(234*2.5)))
    img_resized.save(f'image_dump\\biggus.png')

    return img_resized, img_path

def scrape(card, page):
    
    if page < 2:
        url = f"https://scryfall.com/search?as=grid&order=released&q=%21%22{card}%22+include%3Aextras&unique=prints"
    else:
        url = f"https://scryfall.com/search?as=grid&order=released&page={page}&q=%21%22{card}%22+include%3Aextras&unique=prints"
        

    try:
        driver.get(url)
    except:
        return
    


    with open("website_dump\\scryfall.txt", 'w', encoding="utf-8") as file:
        file.write(driver.page_source)
    with open("website_dump\\scryfall.txt", encoding="utf-8") as file:
        img_list = bs(file, "html.parser").findAll('img')
    
    url_list = []
    
    count = 0
    for img in img_list:
        count += 1

        img_str = str(img)
        start = img_str.rfind('src="') + 5
        end = img_str.find('" title')

        img_url = img_str[start:end]
        if "back" in img_url:
            count -= 1
            continue
        url_list.append(img_url)

    return count, url_list



img_small_col = [[ui.Button(key="IMAGE_SMALL_1", visible=False), 
                  ui.Button(key="IMAGE_SMALL_2", visible=False),
                  ui.Button(key="IMAGE_SMALL_3", visible=False),
                  ui.Button(key="IMAGE_SMALL_4", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_5", visible=False),
                  ui.Button(key="IMAGE_SMALL_6", visible=False),
                  ui.Button(key="IMAGE_SMALL_7", visible=False),
                  ui.Button(key="IMAGE_SMALL_8", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_9", visible=False),
                  ui.Button(key="IMAGE_SMALL_10", visible=False),
                  ui.Button(key="IMAGE_SMALL_11", visible=False),
                  ui.Button(key="IMAGE_SMALL_12", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_13", visible=False),
                  ui.Button(key="IMAGE_SMALL_14", visible=False),
                  ui.Button(key="IMAGE_SMALL_15", visible=False),
                  ui.Button(key="IMAGE_SMALL_16", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_17", visible=False),
                  ui.Button(key="IMAGE_SMALL_18", visible=False),
                  ui.Button(key="IMAGE_SMALL_19", visible=False),
                  ui.Button(key="IMAGE_SMALL_20", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_21", visible=False),
                  ui.Button(key="IMAGE_SMALL_22", visible=False),
                  ui.Button(key="IMAGE_SMALL_23", visible=False),
                  ui.Button(key="IMAGE_SMALL_24", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_25", visible=False),
                  ui.Button(key="IMAGE_SMALL_26", visible=False),
                  ui.Button(key="IMAGE_SMALL_27", visible=False),
                  ui.Button(key="IMAGE_SMALL_28", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_29", visible=False),
                  ui.Button(key="IMAGE_SMALL_30", visible=False),
                  ui.Button(key="IMAGE_SMALL_31", visible=False),
                  ui.Button(key="IMAGE_SMALL_32", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_33", visible=False),
                  ui.Button(key="IMAGE_SMALL_34", visible=False),
                  ui.Button(key="IMAGE_SMALL_35", visible=False),
                  ui.Button(key="IMAGE_SMALL_36", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_37", visible=False),
                  ui.Button(key="IMAGE_SMALL_38", visible=False),
                  ui.Button(key="IMAGE_SMALL_39", visible=False),
                  ui.Button(key="IMAGE_SMALL_40", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_41", visible=False),
                  ui.Button(key="IMAGE_SMALL_42", visible=False),
                  ui.Button(key="IMAGE_SMALL_43", visible=False),
                  ui.Button(key="IMAGE_SMALL_44", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_45", visible=False),
                  ui.Button(key="IMAGE_SMALL_46", visible=False),
                  ui.Button(key="IMAGE_SMALL_47", visible=False),
                  ui.Button(key="IMAGE_SMALL_48", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_49", visible=False),
                  ui.Button(key="IMAGE_SMALL_50", visible=False),
                  ui.Button(key="IMAGE_SMALL_51", visible=False),
                  ui.Button(key="IMAGE_SMALL_52", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_53", visible=False),
                  ui.Button(key="IMAGE_SMALL_54", visible=False),
                  ui.Button(key="IMAGE_SMALL_55", visible=False),
                  ui.Button(key="IMAGE_SMALL_56", visible=False)],
                  [ui.Button(key="IMAGE_SMALL_57", visible=False),
                  ui.Button(key="IMAGE_SMALL_58", visible=False),
                  ui.Button(key="IMAGE_SMALL_59", visible=False),
                  ui.Button(key="IMAGE_SMALL_60", visible=False)]]
                  

img_big_col = [[ui.Image(size=(int(168*2.5), int(234*2.5)), key='IMAGE_LARGE', visible=True)]]

layout = [[ui.Input(size=(200, 1), key='INPUT')],
          [ui.Column(img_small_col, expand_x=True, expand_y=True, scrollable=True, vertical_scroll_only=True, key='COLUMN'),
            ui.Button("Pick", key='PICK', size=(8,4)),
            ui.Column(img_big_col, element_justification='r')],
          [ui.Button("Page 1", key='BACK', size=(20, 3), visible=False), ui.Button("Page 2", key='NEXT', size=(20, 3), visible=False)],
          [ui.Push(), ui.Button("Exit", key='EXIT', size=(20, 3))]]

window = ui.Window("CockSuccor", layout, resizable=True, finalize=True)
window.bind('<Return>', '_ENTER')

urls = []
searched_card = ""
selected_card = ""
big_img = 0
current_page = 1

while True:
    event, values = window.read()
    
    if event == 'EXIT' or event == ui.WIN_CLOSED:
        break
    
    if event == '_ENTER':
        window[f'BACK'].update(visible=False)
        window[f'NEXT'].update(visible=False)
        current_page = 1
        index = 1
        searched_card = values['INPUT']
        count, urls = scrape(values['INPUT'], 1)
        images = get_img_sml(count, urls)
        
        for _ in urls:
            window[f'IMAGE_SMALL_{index}'].update(image_filename=images[index-1], visible=True)
            index += 1
            if index == 61:
                break
        for i in range(index, 60):
            window[f'IMAGE_SMALL_{i}'].update(visible=False)

    if event == 'NEXT':
        current_page += 1
        index = 1
        count, urls = scrape(searched_card, current_page)
        images = get_img_sml(count, urls)
        
        for _ in urls:
            window[f'IMAGE_SMALL_{index}'].update(image_filename=images[index-1], visible=True)
            index += 1
            if index == 61:
                break

    while event == 'BACK':
        if (current_page - 1) == 0:
            break
        index = 1
        current_page -= 1
        count, urls = scrape(searched_card, current_page)
        images = get_img_sml(count, urls)
        
        for _ in urls:
            window[f'IMAGE_SMALL_{index}'].update(image_filename=images[index-1], visible=True)
            index += 1
            if index == 61:
                break

    for i in range(1, 60):
        if event == f'IMAGE_SMALL_{i}':
            selected_card = searched_card
            big_img, url = get_img_big(urls[i-1])
            window[f'IMAGE_LARGE'].update(url, visible=True)

    if event == 'PICK' and selected_card != "":
        big_img.save((cock_path + f'\\{selected_card}.png'))

    

window.close()
driver.close()
```
